### Base code

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

## Tarefa 1 - Trazer a média das vendas

```python
# Construindo um Data Frame:

df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
df_faturamento
```

### Média

```python
# Calculando média
df_faturamento.valor.mean()
```

### Mediana

```python
# Calculando mediana
df_faturamento.valor.median()
```

### Moda (aqui não há valor repetido)

```python
# Calculando moda
df_faturamento.valor.mode()
```

## Tarefa 2 - Criar um gráfico de barras vertical

```python
df_faturamento.info()
```

```python
# Convertendo o Dtype de objetc para Date em data_ref
df_faturamento.data_ref = pd.to_datetime(df_faturamento.data_ref)
df_faturamento.info()
```

```python
# Formatando e plotando gráfico vertical
df_faturamento = df_faturamento.sort_values("data_ref")
df_faturamento["ano_mes"] = df_faturamento["data_ref"].dt.to_period("M")

df_faturamento.plot.bar(x='ano_mes', y='valor')
```

## Tarefa 3 - Criar um gráfico de linhas

```python
df_faturamento.plot.line(x='ano_mes', y='valor')
```
